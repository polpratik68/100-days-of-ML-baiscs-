```python
# encoding 
final_df=pd.get_dummies(df)
```

```python
final_df
```

```python
X=final_df.drop('Selling_Price',axis=1)
```

```python
y=final_df['Selling_Price']
```

```python
# scaling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_sc=sc.fit_transform(X)
```

```python
# cross validation 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.2,random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model training

```python
# model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
y_pred1=lr.predict(X_test)
```

```python
# model evaluation 
from  sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
```

```python
# SVM Support vector machine 
```

```python
# model 2 
from sklearn.svm import SVR 
svm=SVR()
svm.fit(X_train,y_train)
```

```python
svm.score(X_test,y_test)
```

```python
# svm perform well on small data set 
```

```python
y_pred2=svm.predict(X_test)
```

```python
from  sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred2))
print(mean_squared_error(y_test,y_pred2))
print(mean_absolute_error(y_test,y_pred2))
```

```python
# Randomforest regressor 
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
y_pred3=rf.predict(X_test)
```

```python
from  sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred3))
print(mean_squared_error(y_test,y_pred3))
print(mean_absolute_error(y_test,y_pred3))
```

```python
# Decision tree 
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

```python
y_pred4=dt.predict(X_test)
```

```python
from  sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred4))
print(mean_squared_error(y_test,y_pred4))
print(mean_absolute_error(y_test,y_pred4))
```

```python
# Ridge regression
from sklearn.linear_model import Ridge
re=Ridge(alpha=0.5)
re.fit(X_train,y_train)
```

```python
re.score(X_test,y_test)
```

```python
y_pred5=re.predict(X_test)
```

```python
print(r2_score(y_test,y_pred5))
print(mean_squared_error(y_test,y_pred5))
print(mean_absolute_error(y_test,y_pred5))
```

```python
# lasso ........... its also usefro prevent overfitting  as well as feature selection 
from sklearn.linear_model import Lasso 
la=Lasso(alpha=0.5)
la.fit(X_train,y_train)
```

```python
la.score(X_test,y_test)
```

```python
y_pred6=la.predict(X_test)
```

```python
print(r2_score(y_test,y_pred6))
print(mean_squared_error(y_test,y_pred6))
print(mean_absolute_error(y_test,y_pred6))
```
