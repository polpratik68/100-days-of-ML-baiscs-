# Imbalance data 
# if the problem is classfication based 

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df=pd.read_csv('creditcard.csv')
df
```

```python
df.shape
```

```python
df['Class'].isnull().sum().sum()
```

```python
df['Class'].value_counts()
```

```python
# here is large imbalance data 
```

```python
# avoid imbalance data we use sampling technique 
```

```python
# Sampling Tachnique 

# 1: oversampler 
# 2: undersampler
# 3: SMOTE
```

```python
# separate x and y 
X=df.iloc[:,:30]
```

```python
y=df['Class']
```

```python
y.value_counts()
```

```python
! pip install imblearn
```

# over sampling 
#Randomly duplicate examples in the minority class.

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()
X_ros,y_ros=ros.fit_resample(X,y)
```

```python
X_ros.shape
```

```python
y_ros.shape
```

```python
y_ros.value_counts()
```

```python
# Undersampling 
#Randomly delete examples in the majority class
```

```python

from imblearn.under_sampling import RandomUnderSampler
rus=RandomUnderSampler()
X_rus,y_rus=rus.fit_resample(X,y)
```

```python
X_rus.shape
```

```python
y_rus.shape
```

```python
y_rus.value_counts()
```

```python
# smote (will create new records by itself)
from imblearn.over_sampling import SMOTE
sm=SMOTE()
X_sm,y_sm=sm.fit_resample(X,y)
```

```python
X_sm.shape
```

```python
y_sm.shape
```

```python
y_sm.value_counts()
```
