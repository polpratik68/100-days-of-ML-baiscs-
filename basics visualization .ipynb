```python
import matplotlib.pyplot as plt 
df1 = df[df['species']=='setosa']
df2 = df[df['species']=='versicolor']
plt.scatter(df1['sepal_length'],df1['sepal_width'],color='r',marker='*')
plt.scatter(df2['sepal_length'],df2['sepal_width'],color='b',marker='*')
plt.show
```

```python
df1 = df[df['species']=='setosa']
df2 = df[df['species']=='versicolor']
plt.scatter(df1['sepal_length'],df1['sepal_width'],color='r',marker='*')
plt.scatter(df2['sepal_length'],df2['sepal_width'],color='g',marker='*')
plt.show()
```
